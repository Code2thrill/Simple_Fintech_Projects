**Monte Carlo Analysis To Estimate Risk**

```python
PYPL.head()
```

```python
starting_price = 217.929993

days = 365

dt = 1 / days

sigma = returns.std()['PYPL']

mu = returns.mean()['PYPL']

def monte_carlo_analysis(starting_price, days, mu, sigma): #mu assocites with the drift and sigma is the volitility of the stock
    price = numpy.zeros(days)
    price[0]=starting_price
    shock = numpy.zeros(days)
    drift = numpy.zeros(days)
    for day in range(1, days): # started with 1 as price[0] as been defined
        shock[day]= numpy.random.normal(loc = mu * dt, scale = sigma * numpy.sqrt(dt))#sqrt= square root
        #random.normal(loc=0.0, scale=1.0, size=None) Draw random samples from a normal (Gaussian) distribution.
        # loc is the Mean (“centre”) of the distribution.
        
        drift[day]= mu * dt
        price[day]=price[day-1]+(price[day-1]*(drift[day]+shock[day]))
    return price #生成了一条365天的线 in an array，有365个加了随机变量的价格

for run in range(100):#生成100条线
    price = monte_carlo_analysis(starting_price, days, mu, sigma)
    
    pyplot.plot(price)
    
    pyplot.xlabel('Days')
    
    pyplot.ylabel('Price')
```

也就是说pypal的价格很大概率是会在$205-$235之间，如果要是价格高于$235属于overvalued, 低于$205可以买入

**Visualize Price Distribution**

```python
number_of_runs=1000

simulations = numpy.zeros(number_of_runs)#save simulation result into an array as dataframe

for run in range(number_of_runs):
    simulations[run] = monte_carlo_analysis(starting_price, days,mu,sigma)[days-1]#最后一天的价格

pyplot.hist(simulations, bins = 100)

pyplot.figtext(0.6,0.8,s ='Starting price: $%.2f' %starting_price)

pyplot.figtext(0.6,0.7,s = 'Mean final price: $%.2f'%simulations.mean())
```

```python
number_of_runs=1000

simulations = numpy.zeros(number_of_runs)#save simulation result into an array as dataframe

quantile = numpy.percentile(simulations, 1)# just 1 quantile

for run in range(number_of_runs):
    simulations[run] = monte_carlo_analysis(starting_price, days,mu,sigma)[days-1]#最后一天的价格

pyplot.hist(simulations, bins = 100)

pyplot.figtext(0.2,0.8,s ='Starting price: $%.2f' %starting_price)

pyplot.figtext(0.2,0.7,s = 'Mean final price: $%.2f'%simulations.mean())

pyplot.figtext(0.2,0.6,s= "Variance (99 percent confidence): $%.2f" %(starting_price - quantile))

pyplot.figtext(0.2,0.5, s= '1 percent quantile: $%.2f' %quantile)

pyplot.axvline(x = quantile, linewidth =3)
```

👆上面的结果有些奇怪,variance = starting price，并且1 percent quantile = 0; the most money we lose at 99% of the time is $0, the quantile evaluates the risk and it shows risks is 0?
